```python

from skimage.morphology import (binary_erosion,
                                binary_dilation,
                                skeletonize,
                                disk,
                                square)

from skimage.segmentation import random_walker

from skimage.filter import sobel, denoise_tv_chambolle

from scipy import ndimage

from skimage.transform import resize
             
#imgCrp = resize(img, array(imgCrp.shape)*2)
imgCrp = img

denoised = binary_erosion(imgCrp, disk(1))
denoised = binary_dilation(denoised, square(2))
edges = sobel(denoised)

denoised = denoise_tv_chambolle(denoised, weight=0.95)

thresh = 0.85
denoised[denoised>thresh] = True
denoised[denoised<thresh] = False

labels, nLabels = ndimage.label(denoised)

ws = random_walker(edges, labels, beta=20, mode='bf', tol=0.05)

edges = sobel(ws)
skeleton =  np.zeros_like(edges, dtype=bool)
skeleton[edges>np.max(edges)/1000.0] = True
skeleton = skeletonize(skeleton)
skeleton = (skeleton+1)%2

b = arange(ws.max() + 1)
shuffle(b)
shuffledLw = b[ws]
shuffledLw[skeleton == 0] = 1;

subplot(2, 2, 1)
imshow(shuffledLw)

subplot(2, 2, 2)
imshow(imgCrp, cmap = cm.gray)
show()
```

```python
from skimage.measure import regionprops

props = regionprops(ws, ['Area'])
area = array([p["Area"] for p in props])

mean = area.mean()
med = median(area)
N = len(area)

_ = hist(area, 13)
hold("on")

_textstr =  "$\mu = %g$\n$M=%g$\n$N=%d$" % (mean, med, N)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

ax = gca().axes
size = 0.3
ax.text(0.95 - size, 0.95, _textstr, transform=ax.transAxes, fontsize=30,
    verticalalignment='top', bbox=props)

ax.set_xlabel("Area (A)", size=40)

show()
```

```python
import mahotas as mh

X = 1


matrices = (
            #Rank 1
            array([[2, X, 2],  
                   [X, X, X],    
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #Rank 4
            array([[2, X, 2],
                   [X, X, 2],     
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, 2]]),
            
            #Rank 4
            array([[2, X, 2],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #+Rot 45
            array([[X, 2, 2],
                   [2, X, X],     
                   [2, X, 2]]),

        
            )


def getBranch(geometryMatrix, skel):
    return mh.morph.hitmiss(skel, geometryMatrix)

def branchedPoints(skel):
    
    _sum = zeros(shape=skel.shape)
    
    for geometryMatrix in matrices:
    
            _sum += getBranch(geometryMatrix, skel)
            
            for i in range(3):
                geometryMatrix = rot90(geometryMatrix)
                _sum += getBranch(geometryMatrix, skel)
        
        
    return _sum

def leastDistance(x, y, tresh=10):
    
    N = len(x)
    doomList = zeros(N, dtype=bool)
    doomList.fill(True)
    
    for i in range(N):
        
        if not doomList[i]:
            continue
            
        xi, yi = x[i], y[i]
        
        for j in range(N):
            
            if i == j:
                continue
            
            xj, yj = x[j], y[j]
            
            rij = sqrt((xi - xj)**2 + (yi - yj)**2)
            
            if rij < tresh:
                doomList[j] = False
    
    return x[doomList], y[doomList]
    
```

```python
from math import atan2, pi

#Creates a NxN matrix whose element i, j is valued by the element's angle from the mid-row.
def angleMatrix(N):
    
    a = zeros(shape=[N, N])

    for i in range(N):
        for j in range(N):

            a[i , j] = atan2(N/2 - i, N/2 - j)*180/pi

    return fliplr(a)
```

```python
def getAllAngles(joints, skeleton, N=10, k=1):

    ang = angleMatrix(N)
    
    My, Mx = skeleton.shape
    
    allAngles = []
    
    for yi, xi in zip(*joints):
        
        if xi < N or yi < N or xi > Mx-N or yi > My-N:
            continue
        
        skellySlice = skeleton[xi - N/2: xi + N/2, 
                         yi - N/2: yi + N/2]
      
        skellySlice[N/2-k:N/2+k, N/2-k:N/2+k] = 0
        
        #subplot(1, 2, 1)
        #imshow(skellySlice, interpolation='nearest')
        skellySlice = binary_dilation(skellySlice, square(k))
        #subplot(1,2,2)
        #imshow(skellySlice, interpolation='nearest')
        #show()
        #return
        
        l = label(skellySlice)
        angles = []
        
        for i in range(1, l.max()):
                  
            angle = ang[where(l==i)].mean()    
            
            angles.append(angle)
            
            
        n = len(angles)
        
        for i in range(n):
            for j in range(i+1, n):
            
                ti = angles[i]%360
                tj = angles[j]%360
                
                trel = max([ti, tj]) - min([ti, tj])
                
                if trel > 180:
                    trel = 360 - trel    
                        
                allAngles.append(trel)
    
    
    return array(allAngles)
```

```python

ax = gca().axes

n = 20
k = 1
nBins = 10

allAngles = getAllAngles(joints, skelly, n, k)

mean = allAngles.mean()
med = median(allAngles)
N = len(allAngles)

a, b, c = ax.hist(allAngles, nBins)
hold("on")

textstr =  "$\mu = %g$\n$M=%g$\n$N=%d$" % (mean, med, N)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

size = 0.3
ax.set_ylim([0, 1.5*a.max()])

ax.text(0.95 - size, 0.95, textstr, transform=ax.transAxes, fontsize=30,
    verticalalignment='top', bbox=props)

ax.set_xlabel("Angle (degrees)", size=40)

show()
```
