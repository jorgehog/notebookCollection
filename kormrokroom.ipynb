```python
res = 1
image = image_orig[::res, ::res, :]
```

```python
imshow(image[:, :, 0])
figure()
imshow(image[:, :, 1])
figure()
imshow(image[:, :, 2])
show()
```

```python
image_red = image[:, :, 0]/256.
image_green = image[:, :, 1]/256.
image_blue = image[:, :, 2]/256.

r_cap = 0.6
g_cap = 0.46
b_cap = 0.3

red_bin = image_red > r_cap
green_bin = image_green > g_cap
blue_bin = image_blue > b_cap
```

```python
imshow(red_bin)
figure()
imshow(green_bin)
figure()
imshow(blue_bin)
show()
```

```python
from skimage import measure

def get_max_region(binary_image):
    
    im_label = measure.label(binary_image)
    
    props = measure.regionprops(im_label)
    
    all_areas = [m.area for m in props]
    max_area = max(all_areas)
    max_index = all_areas.index(max_area)
    
    max_label = props[max_index].label

    I = np.where(im_label==max_label)
    
    return I
```

```python
RI = get_max_region(red_bin)
GI = get_max_region(green_bin)
BI = get_max_region(blue_bin)

new_image = image.copy()
for i in range(3):
    new_image[:, :, i][RI] = 0
    new_image[:, :, i][GI] = 0
    new_image[:, :, i][BI] = 0

imshow(new_image)
show()
```

```python
def get_closest(I, J, region):
    
    dr2 = (region[0] - I)**2 + (region[1] - J)**2
    
    return dr2.min()
```

```python
L, W, _ = image.shape

new_image = zeros_like(new_image)

rgb = [0, 0, 0]
for i in range(L):
    for j in range(W):
        
        rgb[0] = get_closest(i, j, RI)
        rgb[1] = get_closest(i, j, GI)
        rgb[2] = get_closest(i, j, BI)
        
        w = rgb.index(min(rgb))

        new_image[i, j, w] = 255
        
    sys.stdout.flush()
    print "\r%4d/%4d" % (i + 1, L),    
```
