###Processing raw images into efficient binary formats

Now we need to turn the images into Black/White (binary) for efficient processing. The first step is to create cropped (shed the uninteresting parts) grey scaled (don't need the RGB data) image. The crop area is obtained by looking at the above images. The gray image is obtained by transforming the R, G and B values into a single value, i.e. $f(R, G, B) \to X_G$

```python
#A simple weighted rgb to gray converting function.
def rgb2gray(rgb):
    
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    
    gray = (0.2989 * r + 0.5870 * g + 0.1140 * b)/255

    return gray
    
cropBox = [200, 2500, 1000, 3800]
greyImages = []
for i in range(numberOfImages):
    
    #Make the image gray
    imageGrey = rgb2gray(images[i])
    imageGreyCropped = imageGrey[cropBox[0]:cropBox[1], cropBox[2]:cropBox[3]]
    
    greyImages.append(imageGreyCropped)

subplot(2, 2, 1)
imshow(greyImages[0], cmap = greyScale)
title("Grey scale image")
subplot(2, 2, 2)
hist(greyImages[0].flatten(), 256)
title("Grey scale histogram")
show()
```

###Calculating displacement using sequential binary images

Now lets define a function which takes on input an initial slice corner and returns the displacement vector representing this area


```python
#Function for drawing a rectangle on the image display
def addRectangleToPlot(sliceCorner):
    gca().add_patch(Rectangle(sliceCorner, sliceSize, sliceSize, fill=False, color=np.random.rand(3,)))

    
#main displacement calculation function. Takes on input the bottom left point of the slice.
def getDisplacementVector(bwImages, sliceCenter, sliceSize, lookBoxSize, plotResults=False, silent=True):
    
    shape = bwImages[0].shape
    
    #Corners of the reference slice
    x0, y0 = [x_i - sliceSize/2 for x_i in sliceCenter]
    x1, y1 = x0 + sliceSize, y0+sliceSize
    
    #Corners of the lookBox
    X0, Y0 = [x_i - lookBoxSize/2 for x_i in sliceCenter]
    X1, Y1 = X0 + lookBoxSize, Y0 + lookBoxSize 
    
    if X0 < 0 or X1 > shape[0] or Y0 < 0 or Y1 > shape[1]:
        print "center @ %s is out of bounds of picture. lookboxSize too big." % sliceCenter
        print "lookBox: [%d - %d, %d - %d], shape: %s" % (X0, X1, Y0, Y1, str(shape))
        return 
    elif sliceSize >= lookBoxSize:
        print "lookBox must be greater than the slice."
        return
    
    
    #Extract the reference (first) slice
    referenceSlice = bwImages[0][x0:x1, y0:y1] 
    
    xyPairs = np.zeros(shape=(len(bwImages), 2))
    xyPairs[0, 0] = x0
    xyPairs[0, 1] = y0

    #Loop over the remaining images and their index
    for i, image in enumerate(bwImages[1:]):
  
        lookBox = image[X0:X1, Y0:Y1] 
        
        #Extract the position (anchor) of the original slice in the current image
        result = match_template(lookBox, referenceSlice)

        x, y = np.unravel_index(np.argmax(result), result.shape)
        
        
        #Add the new pair to the list
        xyPairs[i+1, 0] = x + X0
        xyPairs[i+1, 1] = y + Y0
        
        if not silent:
            print "match @", xyPairs[i+1], "(original @", xyPairs[0], ")"

    #Subtract the previous slice position
    displacementVectors = xyPairs[1:] - xyPairs[:-1] 
    
    if plotResults:
    
        subplot(2, 3, 1)
        
        #Draw the lookBox on the original bw image
        imshow(bwImages[0], cmap = greyScale)
        gca().add_patch(Rectangle([Y0, X0], lookBoxSize, lookBoxSize, fill=False)) 
        gca().add_patch(Rectangle([y0, x0], sliceSize, sliceSize, fill=False)) 
        title("Original image showing the current lookBox around the given slice")
        
        
        subplot(2,3,2)
        #Draw the original BW image
        imshow(bwImages[0][X0:X1, Y0:Y1], cmap = greyScale)
        hold("on")
        
        #Draw rectangles for every matched shape
        for i, pair in enumerate(xyPairs):
            text(pair[1] + sliceSize/2 - Y0, pair[0] + sliceSize/2 - X0, str(i)) #add digits to the plot
    
            addRectangleToPlot([pair[1] - Y0, pair[0] - X0])
            
        #Set the limits to be inside the lookBox
        title("Slice traces (in lookBox)")
        xlim([0, lookBoxSize])
        ylim([lookBoxSize, 0])
        
        
        subplot(2,3,3)
        hold("on")
        #Draw the displacementVectors
        old = array([0,0])
        for i in range(len(bwImages)-1):
            new = old + displacementVectors[i]
            plot([old[0], new[0]], [old[1], new[1]])
            old = new.copy()
        title("Displacement vectors")
    
        show()
    
    return displacementVectors

```

The function takes on input:
<ul>

    <li> The size of the square reference slice (pixel length) </li>
    <li> The size of the lookBox, i.e. the box in which the displaced image is assumed to be contained (centered around the slice). </li>
    <li> An option to display the results. </li>

###Calculating the strain

The <i>displacement gradient matrix</i> (DGM) is defined (in two dimensions) as

$$
\mathrm{DGM} \equiv \begin{bmatrix}
       \frac{\partial u_1}{\partial x_1} & \frac{\partial u_1}{\partial x_2}  \\\
       \frac{\partial u_2}{\partial x_1} & \frac{\partial u_2}{\partial x_2}
\end{bmatrix},
$$

where $x_i$ denotes a principal direction and $u_i$ is the displacement vector along this direction.

The DGM is closely connected to the <i>finite deformation strain tensor</i> in the following manner:

$$
\gamma_{ij} = \frac{1}{2}\left(\frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i}\right) + \underbrace{\frac{1}{2}\sum_k \left(\frac{\partial u_k}{\partial x_i}\frac{\partial u_k}{\partial x_j}\right)}_{\simeq 0 \text{ for small displacements}}.
$$

The strain can then efficiently be calculated as 

$$
\epsilon_{ij} = \frac{1}{2}\left(\frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i}\right).
$$

In order to evaluate the partial differentials, we need a discretized approximation to the true vector field of displacements. 

We achieve this by defining a grid, calculating the average displacement in every grid point by the use of the function introduced above.  

```python
def calculateDisplacementVectorField(bwImages, resolution, initialBoundarySpacing, cancelDrift = True, plotField=False, silent=True):
    """
        This function calculates the displacement vector field at a uniformly spaced mesh with resulution
    [resolution]. A cut of [initialBoundarySpacing] is done around the edges of the original pictures.
    """
    
    
    #Calculate the distance between each meshpoint
    shape = bwImages[0].shape
    xRes, yRes = resolution
    
    dx = int(shape[0] - 2*initialBoundarySpacing )/(xRes - 1)
    dy = int(shape[1] - 2*initialBoundarySpacing )/(yRes - 1)
    
    gridPointsX = np.arange(initialBoundarySpacing, shape[0]-initialBoundarySpacing + 1, dx)
    gridPointsY = np.arange(initialBoundarySpacing, shape[1]-initialBoundarySpacing + 1, dy)
    
    if not silent:
        print "%d, %d, %d" % (initialBoundarySpacing, shape[0] - initialBoundarySpacing + 1, dx)
        print "%d, %d, %d" % (initialBoundarySpacing, shape[1] - initialBoundarySpacing + 1, dy)
        
        print "Original shape: ", shape
        print "X from %d to %d with %d values" % (gridPointsX[0], gridPointsX[-1], len(gridPointsX))
        print "Y from %d to %d with %d values" % (gridPointsY[0], gridPointsY[-1], len(gridPointsY))
        print "DX: %d | DY: %d" % (dx, dy)
    
    
    #We store displacement vector for every point in our descrete space
    vectorField = np.zeros(shape=(xRes, yRes, 2))
    
    #Calculate the total drift in case this is sought cancelled
    _totalDrift = np.zeros(2)
    for i, x in enumerate(gridPointsX):
        for j, y in enumerate(gridPointsY):
            displacementVector = getDisplacementVector(bwImages, [x, y], 50,100 , plotResults=not silent)[1]
            
            vectorField[i, j, :] = displacementVector #sum(displacementVector)/len(displacementVector)
            _totalDrift += displacementVector
        
    _avgDrift = _totalDrift/(xRes*yRes)

    if cancelDrift:
        for i in range(xRes):
            for j in range(yRes):
                vectorField[i, j, :] -= _avgDrift;
        _totalDrift = zeros(2)
        for i in range(xRes):
            for j in range(yRes):
                _totalDrift += vectorField[i, j, :]
        
        print "Cancelled total linear drift.",
        
    print "Resulting drift:", _totalDrift

        
    if plotField:
        subplot(2, 2, 1)
        
        imshow(bwImages[0], cmap = greyScale)
        
        #Make X and Y axis for quiver which inverts the order to match imshow
        X, Y = meshgrid(gridPointsY, gridPointsX)
        hold("on")

        #Apply the same iversion to the vector field.
        U, V = vectorField[:, :, 0], vectorField[:, :, 1]
        quiver(X, Y, V, np.flipud(U))
        
        #Make the axis imitate that of imshow (i.e. inverted x on the y)
        y0, y1 = ylim()
        #ticks =np.linspace(y0, y1, 5) 
        #yticks(ticks, [int(x) for x in ticks[::-1]])
        
        show()
    
    return vectorField, dx, dy


    
    
```

Start calculating the strain



```python

xRes, yRes = resolution

epsTot = np.zeros(shape=(2,2))

eps = np.zeros(shape=(2,2))
e2 = eps.copy()

X = 0
Y = 1

trM = trM2 = 0

for i in range(1, xRes-1):
    for j in range(1, xRes-1):
        
        #u(x, y)
        u = field[i, j]
        
        #u(x, y + dy)
        u_p_y = field[i, j+1]
        
        #u(x, y - dy)
        u_m_y = field[i, j-1]
        
        #u(x + dx, y)
        u_p_x = field[i+1, j]
        
        #u(x - dx, y)
        u_m_x = field[i-1, j]

        #Using two-point 'leapfrog' method (1D crank-nichelson)
        #f'(x) = [ f(x + dx) - f(x - dx) ] / (2 dx) + O(dx^2)
        
        du_x_dx =  (u_p_x[X] -  u_m_x[X])/(2*dx)
        du_y_dx =  (u_p_x[Y] -  u_m_x[Y])/(2*dx)
        
        du_y_dy =  (u_p_y[Y] -  u_m_y[Y])/(2*dy)
        du_x_dy =  (u_p_y[X] -  u_m_y[X])/(2*dy)
        
        #Seting up the strainMatrix
        eps[X, X] = du_x_dx
        eps[Y, X] = eps[X, Y] = 0.5*(du_x_dy + du_y_dx) 
        eps[Y, Y] = du_y_dy 
        
        epsTot += eps
        e2 += eps*eps
        
        tr = eps[X, X] + eps[Y, Y]
        
        trM += tr
        trM2 += tr*tr

        

N = (xRes-2)*(yRes-2)

epsTot /= N
e2 /= N

trM/=N
trM2/=N


print "Trace average:\n %g +- %g\n" % (trM, sqrt(trM2 - trM*trM))


print "Matrix average:\n"
print epsTot
print "\n + - \n"
print np.sqrt(e2 - epsTot*epsTot)
```
