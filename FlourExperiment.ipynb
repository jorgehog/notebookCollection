```python
import mahotas as mh
import itertools

X = 1



matrices = (
            #Rank 1
            array([[2, X, 2],  
                   [X, X, X],    
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #Rank 4
            array([[2, X, 2],
                   [X, X, 2],     
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, 2]]),
            
            #Rank 4
            array([[2, X, 2],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #+Rot 45
            array([[X, 2, 2],
                   [2, X, X],     
                   [2, X, 2]]),

        
            )


def getBranch(geometryMatrix, skel):
    return mh.morph.hitmiss(skel, geometryMatrix)

def branchedPoints(skel):
    
    _sum = zeros(shape=skel.shape)
    
    for geometryMatrix in matrices:
    
            _sum += getBranch(geometryMatrix, skel)
            
            for i in range(3):
                geometryMatrix = rot90(geometryMatrix)
                _sum += getBranch(geometryMatrix, skel)
        
        
    return _sum

def leastDistance(x, y, tresh=10):
    
    N = len(x)
    doomList = zeros(N, dtype=bool)
    doomList.fill(True)
    
    for i in range(N):
        
        if not doomList[i]:
            continue
            
        xi, yi = x[i], y[i]
        
        for j in range(N):
            
            if i == j:
                continue
            
            xj, yj = x[j], y[j]
            
            rij = sqrt((xi - xj)**2 + (yi - yj)**2)
            
            if rij < tresh:
                doomList[j] = False
    
    return x[doomList], y[doomList]
    
```

```python
def getAllAngles(joints, skeleton, N=10, k=1):

    ang = angleMatrix(N)
    
    My, Mx = skeleton.shape
    
    allAngles = []
    
    for yi, xi in zip(*joints):
        
        if xi < N or yi < N or xi > Mx-N or yi > My-N:
            continue
        
        skellySlice = skeleton[xi - N/2: xi + N/2, 
                         yi - N/2: yi + N/2]
      
        skellySlice[N/2-k:N/2+k, N/2-k:N/2+k] = 0
        
        #subplot(1, 2, 1)
        #imshow(skellySlice, interpolation='nearest')
        skellySlice = binary_dilation(skellySlice, square(k))
        #subplot(1,2,2)
        #imshow(skellySlice, interpolation='nearest')
        #show()
        #return
        
        l = label(skellySlice)
        angles = []
        
        for i in range(1, l.max()):
                  
            angle = ang[where(l==i)].mean()    
            
            angles.append(angle)
            
            
        n = len(angles)
        
        for i in range(n):
            for j in range(i+1, n):
            
                ti = angles[i]%360
                tj = angles[j]%360
                
                trel = max([ti, tj]) - min([ti, tj])
                
                if trel > 180:
                    trel = 360 - trel    
                        
                allAngles.append(trel)
    
    
    return array(allAngles)
```

```python
fig, axes = subplots(1, 3, figsize=(35,10))

histSize = {'10mm' : 20, '21mm' : 20, '40mm' : 10}
ks =       {'10mm' : 1,  '21mm' : 2,  '40mm' : 4}
ns =       {'10mm' : 10, '21mm' : 20, '40mm' : 15}

for experiment, ax in zip(subExperiments.keys(), axes):
    
    joints = images[experiment + 'joints']
    sk = images[experiment + 'skelly']
    
    n = ns[experiment]
    k = ks[experiment]
    
    allAngles = getAllAngles(joints, sk, n, k)
    
    mean = allAngles.mean()
    med = median(allAngles)
    N = len(allAngles)
    
    a, b, c = ax.hist(allAngles, histSize[experiment])
    hold("on")
    
    textstr =  "$\mu = %g$\n$M=%g$\n$N=%d$" % (mean, med, N)
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    
    size = 0.3
    ax.set_ylim([0, 1.5*a.max()])
    ax.text(0.95 - size, 0.95, textstr, transform=ax.transAxes, fontsize=30,
        verticalalignment='top', bbox=props)
    
    ax.set_xlabel("Angle (degrees)", size=40)
    ax.set_title(experiment, size=50)

show()
    
```
