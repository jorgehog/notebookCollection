```python
fig, axes = subplots(1, 3, figsize=(35,10))

axes[0].set_ylabel("N(area)", size=40)

cuts = {'10mm' : 7000,
        '21mm' : 22500,
        '40mm' : 50000}

bins = {'10mm' : 25,
        '21mm' : 18,
        '40mm' : 13}

for experiment, ax in zip(subExperiments.keys(), axes):

    props = regionprops(images[experiment + "labels"], ['Area'])
    area = array([p["Area"] for p in props])
    
    area = area[where(area < cuts[experiment])]
    
    mean = area.mean()
    med = median(area)
    N = len(area)
    
    _ = ax.hist(area, bins[experiment], normed=True)
    hold("on")
    
    xMax = ax.get_xlim()[1]
    yMax = ax.get_ylim()[1]
    
    textstr =  "$\mu = %g$\n$M=%g$\n$N=%d$" % (mean, med, N)
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    
    size = 0.3
    ax.text(0.95 - size, 0.95, textstr, transform=ax.transAxes, fontsize=30,
        verticalalignment='top', bbox=props)
    
    ax.set_xlabel("Area (A)", size=40)
    ax.set_title(experiment, size=50)

            
    draw()
show()


```
