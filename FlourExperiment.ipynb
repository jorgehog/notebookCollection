```python
import mahotas as mh

X = 1


matrices = (
            #Rank 1
            array([[2, X, 2],  
                   [X, X, X],    
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #Rank 4
            array([[2, X, 2],
                   [X, X, 2],     
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, 2]]),
            
            #Rank 4
            array([[2, X, 2],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #+Rot 45
            array([[X, 2, 2],
                   [2, X, X],     
                   [2, X, 2]]),

        
            )


def getBranch(geometryMatrix, skel):
    return mh.morph.hitmiss(skel, geometryMatrix)

def branchedPoints(skel):
    
    _sum = zeros(shape=skel.shape)
    
    for geometryMatrix in matrices:
    
            _sum += getBranch(geometryMatrix, skel)
            
            for i in range(3):
                geometryMatrix = rot90(geometryMatrix)
                _sum += getBranch(geometryMatrix, skel)
        
        
    return _sum

def leastDistance(x, y, tresh=10):
    
    N = len(x)
    doomList = zeros(N, dtype=bool)
    doomList.fill(True)
    
    for i in range(N):
        
        if not doomList[i]:
            continue
            
        xi, yi = x[i], y[i]
        
        for j in range(N):
            
            if i == j:
                continue
            
            xj, yj = x[j], y[j]
            
            rij = sqrt((xi - xj)**2 + (yi - yj)**2)
            
            if rij < tresh:
                doomList[j] = False
    
    return x[doomList], y[doomList]
    
```
