For starters, we're going need some image analysis tools and some standard Python packages:

```python
from skimage import img_as_ubyte
from skimage.io import imread
from skimage.morphology import label, watershed, disk, opening, square, dilation, erosion
from skimage.filter import rank

from skimage.measure import regionprops

from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries

from scipy import ndimage

from os.path import join

from matplotlib.cm import Greys_r as greyScale

rcParams['figure.figsize'] = (20.0, 16.0)

gimshow = lambda im: imshow(im, cmap = greyScale)
```

Setting the path to where the data is stored (locally)

```python
mainPath = '/home/jorgehog/fysGeo/Strain/group1'

#These are the folder names of the different experiments
#The list values indicate the range of the images used in the experiment
subExperiments = {'10mm' : [45, 52], '21mm' : [20, 29], '40mm' : [32, 41]}
```

Setting the parameters for the current session

```python
images = {}

cropBox = [760, 2450, 1100, 3100]
rotation = -7 #degrees

for thisExperiment in subExperiments.keys():
    imSpan = subExperiments[thisExperiment]
    imagePaths = [join(mainPath, thisExperiment, "Strain00%d.tif" % i) for i in range(imSpan[0], imSpan[1]+1)]
   
    images[thisExperiment] = []
    for imageFile in imagePaths:
    
        image = img_as_ubyte(imread(imageFile, dtype=uint8, as_grey=True))
        
        image = ndimage.rotate(image, rotation)
        image = image[cropBox[0]:cropBox[1], cropBox[2]:cropBox[3]]
        
        images[thisExperiment].append(image)

    print "The following files were assigned to the %s experiment:\n" % thisExperiment
    for path in imagePaths:
        print "\t", path
    print

subplot(2, 2, 1)
imshow(images['40mm'][-1], cmap = greyScale)
title("Grey scale image")
subplot(2, 2, 2)
hist(images['40mm'][-1].flatten(), 256)
title("Grey scale histogram")
show()
```

Processing the 10mm image

```python
experiment = "10mm"
exImages = images[experiment]

denoised = rank.gradient(exImages[-1], disk(6))
denoised = dilation(denoised, square(6))
markers = rank.gradient(denoised, disk(5)) < 10
markers = ndimage.label(markers)[0]

#local gradient
denoised = dilation(denoised, square(6))
denoised = erosion(denoised, square(10))
gradient = rank.gradient(denoised, disk(2))

images[experiment + "denoised"] = denoised
images[experiment + "markers"] = markers
images[experiment + "gradient"] = gradient
```
