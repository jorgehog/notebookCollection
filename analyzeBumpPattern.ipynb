#Extract scale from known sizes

```python
g = origImage[940:1110, 1005:1182]
imshow(g)

print "The shape of the image displaying the hole is %s" % str(g.shape)
```

The diameter of the drilled holes were $d=6$ cm. The area of the above slice is thus expected to be $d^2 = 36\mathrm{cm}^2$. 

Given that the image consist of $N\times M$ pixels, the scale can be calculates by the following equation

\begin{align}
N\cdot M \,\,\mathrm{pixels}^2 &= 36 \,\,\mathrm{cm}^2 \\
1 \,\, \mathrm{pixels} &= \,\, \sqrt{\frac{36}{N\cdot M} \mathrm{cm}^2} \\
                       &= \,\, \sqrt{\frac{3600}{N\cdot M} }\mathrm{mm} \\
                       &\equiv \alpha
\end{align}

A second way of estimating is an average of the two directions

\begin{align}
N\,\,\mathrm{pixels} &= 60\,\,\mathrm{mm} \\
M\,\,\mathrm{pixels} &= 60\,\,\mathrm{mm} \\
1\,\, \mathrm{pixels} &= \frac{60}{2}\left(1/M + 1/N\right)\mathrm{mm} \\
                      &= 30\left(1/M + 1/N\right)\mathrm{mm} \\
                      &\equiv \beta
\end{align}

A good estimate for the scale $\gamma$ is thus

$$
\gamma = \frac{1}{2}\left(\alpha + \beta\right)
$$

From the previous cell it is apparent that $N=170$ and $M=177$ giving the following scale

```python
N, M = g.shape[0:2]
p1 = sqrt(3600./(N*M))
p2 = 30*(1./M + 1./N)

pixScale = 0.5*(p1  + p2)
print "scale is 1 pixel to %g mm" % pixScale
```

To control check the scale, the shape of the original image in meters is calculated

```python
No, Mo = [shape*pixScale/1000 for shape in origImage.shape[0:2][::-1]]

print "Accoring to the given scale, the shape of the original image is %g m times %g m" % (No, Mo)
```

This result makes perfect sence.

#Image processing to skeleton shape

```python
from skimage.filter import sobel
from skimage.morphology import (disk,  
                                square, 
                                binary_erosion,
                                binary_dilation,
                                skeletonize,
                                label)

#removes the orientation anchor
mainImage[1900:, 2600:] = 0

#Removes the 'fill'
edges = sobel(mainImage)

skeleton = binary_erosion(mainImage, disk(1)) #erode = slimmer
skeleton = binary_dilation(skeleton, disk(3)) #dilate = fatter
skeleton = skeletonize(skeleton)

imshow(skeleton, cmap = cm.Greys)
savefig('/tmp/skelly.png')
show()
```

#extracting junctions

```python
import mahotas as mh

X = 1


matrices = (
            #Rank 1
            array([[2, X, 2],  
                   [X, X, X],    
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #Rank 4
            array([[2, X, 2],
                   [X, X, 2],     
                   [2, X, 2]]),
            
            #+Rot 45
            array([[X, 2, X],
                   [2, X, 2],
                   [X, 2, 2]]),
            
            #Rank 4
            array([[2, X, 2],
                   [2, X, 2],
                   [X, 2, X]]),
            
            #+Rot 45
            array([[X, 2, 2],
                   [2, X, X],     
                   [2, X, 2]]),
            
            #Rank 4 turn
            array([[2, 2, 2],
                   [2, X, X],     
                   [2, X, 2]]),
            
            #+Rot 45
            array([[2, 2, 2],
                   [2, X, 2],     
                   [X, 2, X]]),
            
            
            )


def getBranch(geometryMatrix, skel):
    return mh.morph.hitmiss(skel, geometryMatrix)

def branchedPoints(skel):
    
    _sum = zeros(shape=skel.shape)
    
    for geometryMatrix in matrices:
    
            _sum += getBranch(geometryMatrix, skel)
            
            for i in range(3):
                geometryMatrix = rot90(geometryMatrix)
                _sum += getBranch(geometryMatrix, skel)
        
        
    return _sum

def leastDistance(x, y, tresh=10):
    
    N = len(x)
    doomList = zeros(N, dtype=bool)
    doomList.fill(True)
    
    for i in range(N):
        
        if not doomList[i]:
            continue
            
        xi, yi = x[i], y[i]
        
        for j in range(N):
            
            if i == j:
                continue
            
            xj, yj = x[j], y[j]
            
            rij = sqrt((xi - xj)**2 + (yi - yj)**2)
            
            if rij < tresh:
                doomList[j] = False
    
    return x[doomList], y[doomList]
    
```

```python
jointMap = branchedPoints(skeleton2)

joints = where(jointMap.transpose() != 0)
joints = leastDistance(*joints, tresh=20)

scatter(*joints, s=20)
hold("on")
imshow(np.logical_not(skeleton2), cmap = cm.gray)
show()
```

#Removing junctions to separate lines for direction analysis

```python
noJointImage = skeleton.copy()

dx = 20

for y, x in zip(*joints):
    noJointImage[x-dx:x+dx, y-dx:y+dx] = 0

labeledLines = label(noJointImage)

b = arange(labeledLines.max() + 1)
shuffle(b)
shuffledLw = b[labeledLines]

counts = np.bincount(shuffledLw.flatten())
_max = np.argmax(counts)


prevMax = shuffledLw.max()
prevMaxI = where(shuffledLw==prevMax)

prevCountMax = where(shuffledLw==_max)

shuffledLw[prevCountMax] = prevMax
shuffledLw[prevMaxI] = _max

imshow(shuffledLw, cmap=cm.gist_earth)
colorbar()
show()
```

#Extract region properties such as length and orientation

```python
from skimage.measure import regionprops

props = regionprops(labeledLines, [
    'Orientation',
    'MajorAxisLength',
    'Area'
])

aTresh = 10

direction = array([((p["Orientation"]+pi)%pi)*180/pi for p in props])
length = array([p["MajorAxisLength"]*pixScale/10 for p in props if p['Area'] > aTresh])
```

#Visualize the results

```python
rcParams["xtick.labelsize"] = 30
rcParams["xtick.major.size"] = 30
rcParams["xtick.minor.size"] = 30
rcParams["ytick.labelsize"] = 30
rcParams["ytick.major.size"] = 30
rcParams["ytick.minor.size"] = 30

fig, axes = subplots(1, 2, figsize=(25, 10))
ax1, ax2 = axes

cuts = [0, 0]

bins = [30 , 30]
_xlabel = ["Orientation [deg]", "Length [cm]"]

for i, area, ax in zip([0, 1], [direction, length], axes):
    
    
    mean = area.mean()
    med = median(area)
    N = len(area)
    
    _ = ax.hist(area, bins[i])
    hold("on")
    
    xMax = ax.get_xlim()[1]
    yMax = ax.get_ylim()[1]
    
    textstr =  "$\mu = %g$\n$M=%g$\n$N=%d$" % (mean, med, N)
    props_plot = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    
    size = 0.3
    ax.text(0.95 - size, 0.95, textstr, transform=ax.transAxes, fontsize=30,
        verticalalignment='top', bbox=props_plot)
    
    ax.set_xlabel(_xlabel[i], size=40)

    
savefig(join(outputGraphicsPath, 'OrientationAndLengthDistributions.png'))
show()
```
