The equilibrium concentration for a three-dimensional surface with no external potential is known for a century to be $C_0 = e^{-3/T^{\ast}}$, where $T^\ast = k_bT/E_b$ is the reduced temperature in units of Boltzmann's constant over a bond energy. 

```python
def c0(T=1.0):
    return exp(-3./T)
```

The concentration $C$ is measured as the number of particles $N$ per volume $V$ at a given time $t$ 

$C(t) = N(t)/V(t),$

and the (relative) supersaturation is defined as

$\Omega(t) = C(t)/C_0 - 1.$

This supersaturation is used in an empirical relationship between the linear growth speed and the supersaturation as follows:

$\Delta h / \Delta t = k \Omega.$

From this we can setup a recurrence relation given a value of the supersaturation:

$h(t + \Delta t) = h(t) + k \Omega(t) \Delta t$

```python
def next_height(k, current_height, current_supersaturation, delta_t):
    return current_height + k*current_supersaturation*delta_t
```

Given that our crystal is confined with a cavity of volume $V(t)$, a change in the crystal height means an opposite change in
the cavity volume:

$V(t + \Delta t) - V(t) = -(h(t + \Delta t) - h(t))A,$

where $A$ is the area of the surface.

This leads to the following recurrence relation for the volume:

$V(t + \Delta t) = V(t) -(h(t + \Delta t) - h(t))A,$

```python
def next_volume(current_volume, current_height, next_height, area):
    return current_volume - (next_height - current_height)*area
```

The only way the surface height can change is if a particle is deposited to or dissolved from the surface.
This is equal to saying that the volume change is equal to the change in the number of particles in solution,
which makes sense because the volume of the solute is constant (it has no means of increasing or decreasing).
We get

$N(t + \Delta t) = N(t) + V(t + \Delta t) - V(t)$

```python
def next_N(current_N, current_volume, next_volume):
    return current_N + (next_volume - current_volume)
```

In order to obtain $N(t)$, we use the previous relationship as follows:

$N(t) = C(t)V(t) = (1 + \Omega(t))C_0V(t)$

```python
def get_N(supersaturation, volume):
    return (1 + supersaturation)*c0()*volume
```

The concentration at the next timestep is

$C(t + \Delta t) = N(t + \Delta t) / V(t + \Delta t)$

```python
def next_concentration(current_supersaturation, current_volume, next_volume):
    
    current_N = get_N(current_supersaturation, current_volume)

    next_N_val = next_N(current_N, current_volume, next_volume) 
    
    return next_N_val/next_volume
```

The supersaturation at the next timestep is thus

$\Omega(t + \Delta t) = C(t + \Delta t)/C_0 - 1$

```python
def next_supersaturation(current_supersaturation, current_volume, next_volume):    
    
    next_concentration_val = next_concentration(current_supersaturation, current_volume, next_volume)
    
    return next_concentration_val/c0() - 1
```

Now, all we need to do is to set initialconditions for the supersaturation $\Omega(0)$, volume $V(0)$ and height $h(0)$, set a time step $\Delta t$ and a number of total steps, area $A$,
and a value of $k$ and $T^\ast$, before we calculate the recurrence relations step by step:

```python
def solve_growth_equations(initial_supersaturation, initial_volume, initial_height, delta_t, steps, area, k, T):
    
    supersaturations = zeros(steps + 1)
    volumes = zeros(steps + 1)
    heights = zeros(steps + 1)
    times = zeros(steps + 1)
    
    supersaturations[0] = initial_supersaturation
    volumes[0] = initial_volume
    heights[0] = initial_height
    
    for step in range(steps):
        times[step+1] = times[step] + delta_t
        heights[step+1] = next_height(k, heights[step], supersaturations[step], delta_t)    
        volumes[step+1] = next_volume(volumes[step], heights[step], heights[step+1], area)
        supersaturations[step+1] = next_supersaturation(supersaturations[step], volumes[step], volumes[step+1])
        
    return times, heights, volumes, supersaturations
```

From my paper we know that $k(T) = e^{-0.14/T}$ for $\Omega >= 0$, and $k(T) = e^{-0.53/T}$ for $\Omega < 0$ else.

```python
def get_k(T, supersaturation):
    
    if (supersaturation < 0):
        w = 0.53
    else:
        w = 0.14
    
    return exp(-w/T)
```

```python
area = 30*30
initial_volume = 20*area

n_runs = 9
initial_supersaturations = linspace(-2./3, 2./3, n_runs)
initial_height = 0
T = 1.0

delta_t = 0.001
steps = 10000

heights = zeros([len(initial_supersaturations), steps+1])
times = zeros_like(heights)
supersaturations = zeros_like(heights)
volumes = zeros_like(heights)

for n, initial_supersaturation in enumerate(initial_supersaturations):
    k = get_k(T, initial_supersaturation) #assume sign of supersat won't change.
    
    times[n, :], heights[n, :], volumes[n, :], supersaturations[n, :] = \
        solve_growth_equations(initial_supersaturation, initial_volume, initial_height, delta_t, steps, area, k, T)
```

#Reverse algorithm

Given an input on the form of the plots above, we want to calculate the rate constant $k$.

This can be done by using the relation

$k = \frac{\partial ^2h}{\partial t \partial \Omega}.$



```python
def get_forward_derivative(f, g):
    return (f[1] - f[0])/(g[1] - g[0])

def get_backward_derivative(f, g):
    return (f[-1] - f[-2])/(g[-1] - g[-2])

def get_derivative(f, g, i):
    return (f[i+1] - f[i-1])/(g[i+1] - g[i-1])
```

```python
def calc_dh_dt(heights, times):
    
    dh_dt = zeros_like(heights)
    
    dh_dt[0] = get_forward_derivative(heights, times)
    dh_dt[-1] = get_backward_derivative(heights, times)
    
    for i in range(1, len(dh_dt) - 1):
        dh_dt[i] = get_derivative(heights, times, i) 
    
    return dh_dt

def calc_df_dsupersaturation(f, supersaturations):
    
    dh_dss = zeros_like(f)
    
    dh_dss[0] = get_forward_derivative(f, supersaturations)
    dh_dss[-1] = get_backward_derivative(f, supersaturations)
    
    for i in range(1, len(f) - 1):
        dh_dss[i] = get_derivative(f, supersaturations, i)
    
    return dh_dss
    
```

```python
def get_k_estimate(heights, supersaturations, times):
    
    dh_dt = calc_dh_dt(heights, times)
    
    k_estimate_vs_t = calc_df_dsupersaturation(dh_dt, supersaturations)
    
    k_estimate = k_estimate_vs_t.mean()

    return k_estimate, k_estimate_vs_t, dh_dt
```

```python
n = -1
k_estimate, k_estimate_vs_t, dh_dt = get_k_estimate(heights[n, :], supersaturations[n, :], times[n, :])

input_k = get_k(T, supersaturations[n, 0])

k_r_err = abs((k_estimate - input_k)/input_k)

print "k estimate: %.5f, k input: %.5f, relative error: %g" % (k_estimate, input_k, k_r_err)

plot(times[n, :], supersaturations[n, :], label=r"$\Omega$")
plot(times[n, :], dh_dt, label=r"$\partial h / \partial t$")
xlabel("Time", fontsize=30)
legend(fontsize=30)

figure()
plot(times[n, :], k_estimate_vs_t, 'k--s')
xlabel("Time", fontsize=30)
ylabel(r"$\partial^2 h / \partial t \partial \Omega$", fontsize=30)
show()
```

```python
k_estimates = []
k_r_errors = []


for n in range(n_runs):
    k_estimate, k_estimate_vs_t, dh_dt = get_k_estimate(heights[n, :], supersaturations[n, :], times[n, :])

    input_k = get_k(T, supersaturations[n, 0])

    k_r_err = abs((k_estimate - input_k)/input_k)
    
    k_estimates.append(k_estimate)
    k_r_errors.append(k_r_err)

ss_highres = linspace(initial_supersaturations[0], initial_supersaturations[-1])    
    
figure()
plot(initial_supersaturations, k_estimates, 'ks', label="estimated")
plot(ss_highres, [get_k(T, ss) for ss in ss_highres], label="input")
xlabel(r"$\Omega$", fontsize=30)
ylabel(r"k", fontsize=30)
legend(loc="upper left", fontsize=30)

figure()
plot(initial_supersaturations, k_r_errors, 'ks')
xlabel(r"$\Omega$", fontsize=30)
ylabel(r"relative error", fontsize=30)

show()
```

```python
def solve(T, k, h0, inital_omega, delta_t, N):
    
    heights = zeros(N)
    omegas = zeros(N)
    
    heights[0] = 0
    omegas[0] = inital_omega
    
    fac = -k*h0*(1 - c0(T)*(1 + inital_omega))/c0(T)
    
    for i in range(0, N-1):
     
        domega = fac*omegas[i]/(h0 - heights[i])**2
        
        omegas[i+1] = omegas[i] + delta_t*domega
        
        dheight = k*omegas[i+1] #kramer
        
        heights[i+1] = heights[i] + delta_t*dheight
    
    return omegas, heights
    
```

```python
def c(k, K, o0, h0):
    return k*o0 + K/h0
```

```python
def speed(c, K, h0, h):
    x = h - h0
    return c + K/x
```

```python
from scipy.special import lambertw

def omegasolution(k, a_val, b_val, c1_val, time):
    g = -a_val**2*time/b_val
    u = c1_val*exp(g)
    wu = lambertw(u)
    
    return (a_val/k)*(1+ 1./u*0)/(1. + 1./wu)
```
