```python
import wowapi
import urllib2
import re
```

```python
html = urllib2.urlopen("http://www.guildaudit.com/g/11328")

pattern = r"/draenor/(\w+)/advanced"
members = re.findall(pattern, html.read())
```

```python
results = []
max_name_length = 0
min_ilvl = 1000
max_ilvl = 0
for member in members:
    resource = wowapi.get_character('eu.battle.net', 
                                    'draenor', 
                                    member,
                                    fields="items")
    
    average_ilvl = resource.items["averageItemLevelEquipped"]
    number_of_crafted = 0
    for item_description, item_data in resource.items.items():
        
        if type(item_data) is not dict:
            continue
        
        if item_data["context"] != "trade-skill":
            continue
        
        is_weapon = (item_description == "mainHand" or item_description == "offHand") 
        ilvl = item_data["itemLevel"]
        
        if ilvl == 715 or (ilvl == 705 and is_weapon):
            number_of_crafted += 1

    results.append([member, number_of_crafted, average_ilvl])
    
    if len(member) > max_name_length:
        max_name_length = len(member)
    if average_ilvl > max_ilvl:
        max_ilvl = average_ilvl
    if average_ilvl < min_ilvl:
        min_ilvl = average_ilvl
```

```python
def dump_data(data):
    for member, number_of_crafted, average_ilvl in data:
        print "%s (%3d) has %1d crafted items." % (member.ljust(max_name_length), 
                                                   average_ilvl,
                                                   number_of_crafted)
```

```python
print "sorted by ilvl:"
dump_data(sorted(results, key=lambda l: l[2])[::-1])

print "\n\nsorted by number of crafted items:"
dump_data(sorted(results, key=lambda l: l[1])[::-1])
```
